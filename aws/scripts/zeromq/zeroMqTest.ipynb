```python
import zmq
import time

def main():
    context = zmq.Context()
    socket = context.socket(zmq.PUB)
    socket.bind("tcp://*:5555")

    print("Server started at tcp://localhost:5555")
    while True:
        # Simulate waiting for new data; in a real app, this could be a database update, user action, etc.
        time.sleep(5)  # Wait for 5 seconds to simulate new data arriving
        print("Sending update...")
        socket.send_string("update: Here's some new data!")

if __name__ == "__main__":
    main()
```
